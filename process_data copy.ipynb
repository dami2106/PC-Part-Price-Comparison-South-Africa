```python
def calculate_similarity(title1, title2):
    title1 = title1.lower().strip()
    title2 = title2.lower().strip()

    punc = '''!()-[]{};:'"\,<>./?@#$%^&*_~'''
 
    for ele in title1:
        if ele in punc:
            title1 = title1.replace(ele, "")

    for ele in title2:
        if ele in punc:
            title2 = title2.replace(ele, "")

    sim_count = 0
    for word in title1.split():
        if word in title2:
            sim_count += 1

    len_ = max(len(title1.split()), len(title2.split()))
    return sim_count/ len_ * 100
```

```python
def get_match(title, cate, df):
    max_sim = -100
    max_sim_title = ''
    getDf = df.loc[df['Category'] == str(cate)]
    for df_title in getDf['Title']:
        sim = calculate_similarity(title, df_title)
        if sim > max_sim:
            max_sim = sim
            max_sim_title = df_title

    return max_sim_title, max_sim
```

```python
count = 0
for i in range(len(dwt)):
    match = get_match(dwt.iloc[i]['Title'], dwt.iloc[i]['Category'], wootware)
    if match[1] < 65:
        continue
    else:
        print("WOOTWARE : " + match[0])
        print("DWT : " + dwt.iloc[i]['Title'])
        print(str(match[1]) + "\n")
        count += 1
```
