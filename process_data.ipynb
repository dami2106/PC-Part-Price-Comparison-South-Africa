```python
def calculate_similarity(title1, title2):
    title1 = title1.lower().strip()
    title2 = title2.lower().strip()

    punc = '''!()-[]{};:'"\,<>./?@#$%^&*_~'''
 
    for ele in title1:
        if ele in punc:
            title1 = title1.replace(ele, "")

    for ele in title2:
        if ele in punc:
            title2 = title2.replace(ele, "")

    sim_count = 0
    for word in title1.split():
        if word in title2:
            sim_count += 1

    len_ = max(len(title1.split()), len(title2.split()))
    return sim_count/ len_ * 100
```

```python
def get_match(title, cate, df):
    max_sim = -100
    max_sim_title = ''
    getDf = df.loc[df['Category'] == str(cate)]
    for df_title in getDf['Title']:
        sim = calculate_similarity(title, df_title)
        if sim > max_sim:
            max_sim = sim
            max_sim_title = df_title

    return max_sim_title, max_sim
```

```python
import pandas as pd
from name_matching.name_matcher import NameMatcher

# define a dataset with computer parts names from wootware
wootware_a = wootware

# define a dataset with computer parts names from dreamwaretech
dwt_a = dwt

# initialise the name matcher
matcher = NameMatcher(number_of_matches=1, 
                      legal_suffixes=True, 
                      common_words=False, 
                      top_n=50, 
                      verbose=True)

# adjust the distance metrics to use
matcher.set_distance_metrics(['bag', 'typo', 'refined_soundex','discounted_levenshtein'])

# load the data to which the names should be matched
matcher.load_and_process_master_data(column='Title',
                                     df_matching_data=wootware_a, 
                                     transform=True)

# perform the name matching on the data you want matched
matches = matcher.match_names(to_be_matched=dwt_a, 
                              column_matching='Title')

# combine the datasets based on the matches
combined = pd.merge(wootware_a, matches, how='left', left_index=True, right_on='match_index')
combined = pd.merge(combined, dwt_a, how='left', left_index=True, right_index=True)
```
